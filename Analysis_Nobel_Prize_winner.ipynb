<a id="load-the-data"></a>
#### Loading data

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Reading in the Nobel Prize data
nobel = pd.read_csv(
    "datasets/nobel.csv")

# Taking a look at the first several winners
nobel.head(6)
```

<a id="intital-data-analysis"></a>
#### Initial data analysis based on individual vs. shared, sex and country

```python
# Display the number of (possibly shared) Nobel Prizes handed
# out between 1901 and 2016
nobel['laureate_type'].value_counts()

# Display the number of prizes won by male and female recipients.
nobel['sex'].value_counts()

# Display the number of prizes won by the top 10 nationalities.
nobel['birth_country'].value_counts().head(10)
```

<a id="USA-Analysis"></a>
#### Analyzing USA data to understand the dominance

```python
# Calculating the proportion of USA born winners per decade
nobel['usa_born_winner'] = nobel['birth_country'] == "United States of America"
nobel['decade'] = (np.floor(nobel['year']/10)*10).astype(int)
prop_usa_winners = nobel.groupby('decade', as_index=False).agg({
    'usa_born_winner': 'mean'})

# Display the proportions of USA born winners per decade
prop_usa_winners
```

<a id="USA-Visualization"></a>
#### Visualization of USA Nobel prize winners as percentage in each decade

```python
# Setting the plotting theme
from matplotlib.ticker import PercentFormatter
import matplotlib.pyplot as plt
sns.set()
# and setting the size of all plots.
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners
ax = sns.lineplot(x='decade', y='usa_born_winner', data=prop_usa_winners)

# Adding %-formatting to the y-axis
ax.yaxis.set_major_formatter(PercentFormatter())
```

<a id="Gender-Analysis-Visualization"></a>
#### Gender based analysis and visualization of Nobel prize winners

```python
# Calculating the proportion of female laureates per decade
from matplotlib.ticker import PercentFormatter
nobel['female_winner'] = nobel['sex'] == 'Female'
prop_female_winners = nobel.groupby(
    ['decade', 'category'], as_index=False).agg({'female_winner': 'mean'})

# Plotting USA born winners with % winners on the y-axis
ax = sns.lineplot(x='decade', y='female_winner',
                  data=prop_female_winners, hue="category")
ax.yaxis.set_major_formatter(PercentFormatter())
```

<a id="First-Female-Winner"></a>
#### Finding the first female nobel prize winner

```python
# Picking out the first woman to win a Nobel Prize
nobel[nobel['sex'] == 'Female'].nsmallest(1, 'year')
```

<a id="People-with-multiple-Nobles"></a>
#### Finding people with multiple noble prizes

```python
# Selecting the laureates that have received 2 or more prizes.
nobel.groupby('full_name').filter(lambda x: x['full_name'].count() >= 2)
```

<a id="Visualizing-age-to-year-of-win"></a>
#### Visualizing age to year of winning Nobel prize

```python
# Converting birth_date from String to datetime
nobel['birth_date'] = pd.to_datetime(nobel['birth_date'])

# Calculating the age of Nobel Prize winners
nobel['age'] = nobel['year']-nobel['birth_date'].dt.year

# Plotting the age of Nobel Prize winners
sns.lmplot(x='year', y='age',data=nobel)
```

<a id="Visualizing-age-to-year-of-win-for-different-categories"></a>
#### Visualizing age to year of winning Nobel prize for different categories

```python
# Same plot as above, but separate plots for each type of Nobel Prize
sns.lmplot(x='year', y='age', data=nobel, row='category')
```

<a id="finding-oldest-and-youngest-winners"></a>
#### Finding oldest and youngest winners

```python
# The oldest winner of a Nobel Prize as of 2016
print(nobel.nlargest(1, 'age'))

# The youngest winner of a Nobel Prize as of 2016
print(nobel.nsmallest(1, 'age'))
```
